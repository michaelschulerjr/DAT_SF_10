```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from patsy import dmatrices
import statsmodels.api as sm

%matplotlib inline
```

```python
df = pd.read_csv("titanitc_train.csv") 
```

```python
#drop out Ticket and Cabin.  Cabin has less than 25% non-null values. Ticket # can be represented by the Class number. 
#For example: when a ticket number begins with the number 3 they are in 3rd class; when ticket number begins with 2 they belong to 2nd class, and so on.
df = df.drop(['Ticket', 'Cabin'], axis=1)
```

```python
# Remove NaN values in order to print graphically. 
df_NaN_removed = df.dropna() 
```

```python
# specifies the parameters of our graphs
fig = plt.figure(figsize=(18,6), dpi=1600) 
alpha=alpha_scatterplot = 0.2 
alpha_bar_chart = 0.55

# lets us plot many diffrent shaped graphs together 
ax1 = plt.subplot2grid((2,3),(0,0))
# plots a bar graph of those who surived vs those who did not.               
df_NaN_removed.Survived.value_counts().plot(kind='bar', alpha=alpha_bar_chart)
# this nicely sets the margins in matplotlib to deal with a recent bug 1.3.1
ax1.set_xlim(-1, 2)
# puts a title on our graph
plt.title("Distribution of Survival, (1 = Survived)")    

plt.subplot2grid((2,3),(0,1))
plt.scatter(df_NaN_removed.Survived, df_NaN_removed.Age, alpha=alpha_scatterplot)
# sets the y axis lable
plt.ylabel("Age")
# formats the grid line style of our graphs                          
plt.grid(b=True, which='major', axis='y')  
plt.title("Survial by Age,  (1 = Survived)")

ax3 = plt.subplot2grid((2,3),(0,2))
df_NaN_removed.Pclass.value_counts().plot(kind="barh", alpha=alpha_bar_chart)
ax3.set_ylim(-1, len(df_NaN_removed.Pclass.value_counts()))
plt.title("Class Distribution")

plt.subplot2grid((2,3),(1,0), colspan=2)
# plots a kernel desnsity estimate of the subset of the 1st class passanges's age
df_NaN_removed.Age[df_NaN_removed.Pclass == 1].plot(kind='kde')    
df_NaN_removed.Age[df_NaN_removed.Pclass == 2].plot(kind='kde')
df_NaN_removed.Age[df_NaN_removed.Pclass == 3].plot(kind='kde')
 # plots an axis lable
plt.xlabel("Age")    
plt.title("Age Distribution within classes")
# sets our legend for our graph.
plt.legend(('1st Class', '2nd Class','3rd Class'),loc='best') 

#shows the distribution of who survived by class
plt.subplot2grid((2,3),(1,2), colspan=1)
# plots a kernel desnsity estimate of the subset of the 1st class passanges's age
df_NaN_removed.Survived[df_NaN_removed.Pclass == 1].plot(kind='kde')    
df_NaN_removed.Survived[df_NaN_removed.Pclass == 2].plot(kind='kde')
df_NaN_removed.Survived[df_NaN_removed.Pclass == 3].plot(kind='kde')
 # plots an axis label
plt.xlabel("Survived")    
plt.title("Survival Distribution within classes")
# sets our legend for our graph.
plt.legend(('1st Class', '2nd Class','3rd Class'),loc='best') 
```

```python
fig = plt.figure(figsize=(18,4), dpi=1600)
alpha_level = 0.65

# building on the previous code, here we create an additional subset with in the gender subset 
# we created for the survived variable. I know, thats a lot of subsets. After we do that we call 
# value_counts() so it it can be easily plotted as a bar graph. this is repeated for each gender 
# class pair.
ax1=fig.add_subplot(141)
female_highclass = df_NaN_removed.Survived[df_NaN_removed.Sex == 'female'][df_NaN_removed.Pclass != 3].value_counts()
female_highclass.plot(kind='bar', label='female highclass', color='#FA2479', alpha=alpha_level)
ax1.set_xticklabels(["Survived", "Died"], rotation=0)
ax1.set_xlim(-1, len(female_highclass))
plt.title("Who Survived? with respect to Gender and Class"); plt.legend(loc='best')

ax2=fig.add_subplot(142, sharey=ax1)
female_lowclass = df_NaN_removed.Survived[df_NaN_removed.Sex == 'female'][df_NaN_removed.Pclass == 3].value_counts()
female_lowclass.plot(kind='bar', label='female, low class', color='pink', alpha=alpha_level)
ax2.set_xticklabels(["Died","Survived"], rotation=0)
ax2.set_xlim(-1, len(female_lowclass))
plt.legend(loc='best')

ax3=fig.add_subplot(143, sharey=ax1)
male_highclass = df_NaN_removed.Survived[df_NaN_removed.Sex == 'male'][df_NaN_removed.Pclass != 3].value_counts()
male_highclass.plot(kind='bar', label='male highclass', alpha=alpha_level, color='steelblue')
ax3.set_xticklabels(["Died","Survived"], rotation=0)
ax3.set_xlim(-1, len(male_highclass))
plt.legend(loc='best')

ax4=fig.add_subplot(144, sharey=ax1)
male_lowclass = df_NaN_removed.Survived[df_NaN_removed.Sex == 'male'][df_NaN_removed.Pclass == 3].value_counts()
male_lowclass.plot(kind='bar', label='male, low class',color='lightblue', alpha=alpha_level)
ax4.set_xticklabels(["Died","Survived"], rotation=0)
ax4.set_xlim(-1, len(male_lowclass))
plt.legend(loc='best')
```

```python
# because there is a different probability on survival depending on which class the person was in and which gender they are, 
# we should view them as 4 different "groups".  We'll divide the Class into highclass and lowclass -- high will include both 
# 1st and 2nd class while low class will include 3rd class.  Then we'll divide Sex by male or female.  The 4 groups will now
# be known as: female_highclass, female_lowclass, male_highclass, male_lowclass.
```
