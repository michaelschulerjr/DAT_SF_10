```python
#import modules
import pandas as pd
from pandas.tools.plotting import scatter_matrix
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
# store csv
data = pd.read_csv('student_logins.csv')
```

```python
#converting objects to datetime

data['session_end_time'] = pd.to_datetime(data['session_end_time'])
data['session_start_time'] = pd.to_datetime(data['session_start_time'])
data['student_account_created'] = pd.to_datetime(data['student_account_created'])
data['Duration'] = data['session_end_time'] - data['session_start_time']
data['Duration'] = data['Duration'].map(lambda x: x.astype('float64')/(1e9*60))
```

```python
data.info()
```

```python
MaxCreated = pd.to_datetime(max(data['session_end_time']))
data['AccountAge'] = MaxCreated - data['student_account_created']
data['AccountAge'] = data['AccountAge'].map(lambda x: x.astype('float64')/(1e9*60*60*24))
```

```python
data[0:20]
```

```python
class_dummies = pd.get_dummies(data['class_id']) 
student_id_dummies = pd.get_dummies(data['student_id'])
class_student_interaction = class_dummies * student_id_dummies
```

```python
#perform OLS regression (1)

df1 = data[(data['Duration'] < 225) & (data['Duration'] > 0)]
df1 = pd.merge(data,class_dummies,left_index=True,right_index=True,how='left')

x = df1[['problems_completed','AccountAge','student_previous_logins_total','student_previous_class_logins','a','c','e','g','m']].values
y = df1['Duration'].values
X = sm.add_constant(x, prepend=True)

results = sm.OLS(y, X).fit()
results.summary()
```

```python
class_dummies_cols = list(class_dummies.columns)
student_id_dummies_cols = list(student_id_dummies.columns)
rawfeatures = ['problems_completed','AccountAge','student_previous_logins_total','student_previous_class_logins']
features = rawfeatures + class_dummies_cols + student_id_dummies_cols
```

```python
#perform OLS regression(2)

df2 = pd.merge(df1,student_id_dummies,left_index=True,right_index=True,how='left')
x2 = df2[features].values
y2 = df2['Duration'].values
X2 = sm.add_constant(x2, prepend=True)
results2 = sm.OLS(y2, X2).fit()
results2.summary()
```
